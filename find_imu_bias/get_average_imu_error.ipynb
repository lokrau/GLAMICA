```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv('imu_data_one_hour.csv')

# Round timestamps down to the nearest second
df['TimeSec'] = df['Timestamp(s)'].astype(int)
df['TimeSec'] = df['TimeSec'] - df['TimeSec'].min()

# Group by each second and compute the mean for each axis
df_avg = df.groupby('TimeSec')[['GyroX(rad/s)', 'GyroY(rad/s)', 'GyroZ(rad/s)']].mean().reset_index()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df_avg['TimeSec'], df_avg['GyroX(rad/s)'], label='Gyro X (avg)')
plt.plot(df_avg['TimeSec'], df_avg['GyroY(rad/s)'], label='Gyro Y (avg)')
plt.plot(df_avg['TimeSec'], df_avg['GyroZ(rad/s)'], label='Gyro Z (avg)')

plt.title('Averaged IMU Gyroscope Data Per Second')
plt.xlabel('Time (s)')
plt.ylabel('Average Angular Velocity (rad/s)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# average of the GyroX, GyroY, and GyroZ columns
df_mean = df[['GyroX(rad/s)', 'GyroY(rad/s)', 'GyroZ(rad/s)']].mean()

#print that
print("Average GyroX:", df_mean['GyroX(rad/s)'])
print("Average GyroY:", df_mean['GyroY(rad/s)'])
print("Average GyroZ:", df_mean['GyroZ(rad/s)'])
```
